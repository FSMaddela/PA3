## Problem 2: Subsetting, Slicing, and IndexingX


**Import pandas as pd and load the csv file**

```python
import pandas as pd

cars = pd.read_csv('cars.csv')
cars
```

**Problem 2.A - Display the first five rows with odd-numbered columns of cars**

```python
odd_cars = cars.iloc[1::2]
odd_cars.head()
```

**Problem 2.B - Display the row that contains the 'Model' of 'Mazda Rx4'**

```python
cars.loc[(cars['Model']=='Mazda RX4')]
```

**Problem 2.C - Display how many cylinders (cyl) the car model Camaro Z28 has**

```python
cars.loc[(cars['Model']=='Camaro Z28'), ['cyl']]
```

**Problem 2.D - Display how many cylinders (cyl) and what gear type (gear) the car models Mazda RX4 Wag, Ford Pantera L and Honda Civic have.**

```python
models = ['Mazda RX4 Wag', 'Ford Pantera L', 'Honda Civic']
cars.loc[cars['Model'].isin(models), ['Model','cyl', 'gear']]
```
